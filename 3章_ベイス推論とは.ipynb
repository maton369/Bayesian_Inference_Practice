```python
# [0, 1]区間一様分布の確率密度関数
def f(x):
    # 関数をユニバーサル関数にするための工夫
    return x - x + 1.0


x = np.arange(0.0, 1.1, 0.1)
plt.fill_between(x, f(x))
plt.title("区間[0, 1]一様分布の確率密度関数")
```
